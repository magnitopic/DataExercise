# Transaction table

```python
import pandas as pd
import numpy as np
df = pd.read_csv("../data/data_norm.csv")
employees = pd.read_csv('../exports/employees.csv')
```

```python
transaction = pd.DataFrame(df[['Total', 'Fecha', 'TipoRegistro', 'MedioPago', 'Observaciones', 'Documento', 'NIF']])
transaction = transaction.reset_index(drop=False)
transaction.columns = ['id', 'TotalPrice', 'Date', 'Type', 'PayMedium', 'Observations', 'Document', 'ClientID']
transaction['id'] += 1

# map employees
employee_id_map = dict(zip(employees['Name'], employees['id']))

transaction['EmployeeID'] = df['Empleado'].map(employee_id_map)
transaction['EmployeeID'] += 1
transaction.head(8)
```

```python
transaction.to_csv('../exports/transaction.csv', index=False)
```
